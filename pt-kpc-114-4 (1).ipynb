4.2. Cache Memory Principles}
#1. word, block
	- [word] is the natural unit of organization of memory
	- 워드의 사이즈는 정수를 표현하는데 쓰이는 비트의 갯수와 같다
	- 워드의 사이즈는 instruction를 표현하는데 쓰이는 비트의 갯수와 같다
		ex. 32bit 컴퓨터에서 1word = 32bit = 4byte, 64bit 컴퓨터에서 1word = 64bit = 8byte
	- 워드의 사이즈가 n이면 메모리의 전체 addressable word는 2^n개이다

	- [block]은 K개의 word를 묶은 단위이다
	- 블락의 사이즈가 K이면 메모리의 전체 block의 갯수는 2^n/K이다. 이를 M이라고 두자
	- Cache에서는 block대신 line이라고 한다. "이때 line에는 tag bit와 control bit를 추가한다"(추후 설명)

### 1. RTOS스케줄링 개요

##### * 실시간 시스템 - 제한된 응답시간이나 오동작을 포함한 결과의 위험성에 관한 명확한 규정을 만족하는 시스템

- 경성 실시간 시스템(Hard Real-time)    
- 준경성 실시간 시스템(Firm Real-time)
- 연성 실시간 시스템(Soft Real-time)


##### * 실시간 스케줄링 - 실시간 운영체제의 스케줄러는 선점을 이용한 우선순위 기반의 스케줄링 알고리즘을 지원해야함

##### * 비율 단조 스케줄링 (rate-monotonic scheduling) - 수행 주기가 가장 짧은 프로세스에 가장 높은 우선순위를 부여하는 방식의 알고리즘

##### *  최단 마감 우선 스케줄링 (Earliest Deadline First Scheduling) - 현재 준비상태에 있는 태스크들 중에서 마감시간이 가장 짧은 태스크를 선택하여 수행




### 2. RM스케줄링 정의, 특징, 동작방식

#### 특징 - 선점형, 프로세스 수행시간 일정, 마감시각은 마지막 주기가 끝나는 시점


#### 충분조건  
- CPU이용율(U) ≤N(2^(1/N)-1)
- 태스크의 개수 N이 증가함에 따라 이용율은 점점 감소하여 69.3%에 수렴함
- 2개 태스크일 경우는 약 83%
- 무한개 태스크일 경우는 약 69%임

#### 동작방식
1. 각 주기 Task들에 대해 우선순위 지정
2. 주기에 반비례하여 우선순위 배정함(주기가 짧으면 높은 우선순위, 길면 낮은 우선순위)
3. 선점가능한 정적 우선순위 정책을 사용하여 주기적인 Task 스케줄링 수행
