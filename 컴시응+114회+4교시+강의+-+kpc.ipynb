### 1. CA/OS 다수 출제  CA 6문제 OS 3문제

### 3. 총31문제 중 CA,NW,보안 24문제


```python
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline

# The slices will be ordered and plotted counter-clockwise.
labels = 'CA', 'OS', 'NW', 'SEC', 'DS', 'IM', 'SE' , 'DB'
sizes = [6, 3, 7, 7, 4, 1, 3, 0]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'lime', 'pink', 'seagreen', 'm']
explode = (0.1, 0.1, 0.1, 0.1 ,0 ,0,0,0) # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')

plt.show()
```
